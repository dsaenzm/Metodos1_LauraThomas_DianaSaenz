```python
#5.2 Espiral de arquímedes

import numpy as np
import matplotlib.pyplot as plt

# En este caso como a= 0 y b=1, significaría que r = ángulo. Al ser polar, x= r*cos(ángulo) y y= r*sin(ángulo). 
angulo = np.linspace(0,2*np.pi, 10000)
x= angulo * np.cos(angulo)
y= angulo * np.sin(angulo)

plt.plot(x,y)

#Si se hace en un rango más grande que (0,2pi), sale más completo

#angulo = np.linspace(0,6*np.pi, 10000)
#x1= angulo * np.cos(angulo)
#y1= angulo * np.sin(angulo)
#plt.plot(x1,y1)

```
